# Análise do Eleitorado Brasileiro

```python
# iportando as libs
import pandas as pd

# lib de graficos
import seaborn as sns
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
# carregar arquivo
df = pd.read_csv("dados/eleitorado_municipio_2018.csv", sep=";", encoding="latin1")


df.head()
```

```python
df.info()
```

```python
# criar variaveis para calcular o percentual dos eleitures
total_eleitores = df["QTD_ELEITORES"].sum()
total_fem = df["QTD_ELEITORES_FEMININO"].sum()
total_masc = df["QTD_ELEITORES_MASCULINO"].sum()
total_nao_informado = df["QTD_ELEITORES_NAOINFORMADO"].sum()
```

```python
# mostrar os valores percentualmente
print("Eleitoras:", ((total_fem / total_eleitores) * 100).round(2), "%")
print("Eleitores:", ((total_masc / total_eleitores) * 100).round(2), "%")
print("Não Informado:", ((total_nao_informado / total_eleitores) * 100).round(2), "%")
```

```python
# quantas municipios com mais homens q mulheres
df[df["QTD_ELEITORES_MASCULINO"] > df["QTD_ELEITORES_FEMININO"]].count()
```

```python
# criar uma coluna nova para indicar a relação fem/masc
df["RELACAO_FM"] = df["QTD_ELEITORES_FEMININO"] / df["QTD_ELEITORES_MASCULINO"]
```

```python
# quais são os municipios com maior relação f/m
df.nlargest(5, "RELACAO_FM")
```

```python
# quais são os municipios com menor relação f/m
df.nsmallest(5, "RELACAO_FM")
```

```python
# gerando um df novo com somente os municipios do Brasil
df_brasil = df[df["CD_PAIS"] == 1].copy()

df_brasil.head()
```

```python
# quais são os municipios com menor relação f/m
df_brasil.nlargest(5, "RELACAO_FM").T
```

```python
# plotar um grafico de distribuição da relação f/m

df_brasil["RELACAO_FM"].plot.hist(bins=100)
```

```python
# plotar a distribuição da relação f/m
# bins => para deixar as linhas mais finas conforme aumenta o valor passado 
# color => definir a cor do grafico
# kdk => remove uma linha por volta do grafico
sns.distplot(df_brasil["RELACAO_FM"], bins=100, color="red", kde=False)

plt.title("Relação Eleitoras/Eleitores", fontsize=16)
plt.xlabel("Eleitoras/Eleitores", fontsize=12)
plt.ylabel("Fequencia", fontsize=12)
plt.axvline(1,0, color="black", linestyle="--")
```

```python
# verificando a distribuição f/m mas mostrando os pontos (municipios) 
# data => de onde vem os dados
# x, y => eixos do grafico
sns.swarmplot(data=df_brasil, x="NM_REGIAO", y="RELACAO_FM")
plt.axhline(1,0, color="black", linestyle="--")
```
